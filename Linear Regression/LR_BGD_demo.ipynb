#### Implement Linear Regression using Batch Gradient Descent (BGD) on Synthetic Data

```python
import numpy as np
import pandas as pd
from LinearRegression_BGD import LinearRegressionBGD
from Linear_Regression_utils import *
```

```python
%matplotlib widget

# Generate synthetic training set
np.random.seed(42)
n_samples = 100
noise = np.random.randn(n_samples) * 15

# Generate y = 3X + 4 + noise
X = np.linspace(1, n_samples, n_samples)
y = 3 * X + 4 + noise

# Fit the model
model = LinearRegressionBGD(learning_rate=0.1, max_iterations=100)
model.fit(X, y)
```

```python
# Plot cost vs. iteration
plot_cost_vs_iteration(model)
```

#### Implement Linear Regression using Batch Gradient Descent (BGD) on Predicting House Price

```python
# Load Data
house_data = pd.read_csv('./dataset/house_price.csv')

# Select first 100 training and next 50 testing data
train_X = np.array(house_data['sqft_living'].head(100))
train_Y = np.array(house_data['price'].head(100))
test_X = np.array(house_data['sqft_living'].iloc[100:150])
test_Y = np.array(house_data['price'].iloc[100:150])

model = LinearRegressionBGD(learning_rate=0.1, max_iterations=100)
model.fit(train_X, train_Y)
```

```python
# Plot cost vs. iteration
plot_cost_vs_iteration(model)

# Evaluate model performance on training data
pred_Y = [model.predict_value(x) for x in test_X]
evaluate_model(model, test_Y, pred_Y)
```
