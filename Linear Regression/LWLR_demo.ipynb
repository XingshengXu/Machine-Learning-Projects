#### Implement Locally Weighted Linear Regression (LWLR) on Synthetic Data

```python
import numpy as np
import pandas as pd
from sklearn.metrics import r2_score
from LWLR import LocalWeightedLinearRegression
```

```python
# Set the random seed for reproducibility
np.random.seed(42)

# Generate 100 sample points
X = np.sort(np.random.rand(100))
y = np.sin(2 * np.pi * X) + 0.3 * np.random.randn(100)

# Generate 100 Query Points from 0 to 1
x_query = np.linspace(0, 1, 100)
```

```python
# Fit the model
model = LocalWeightedLinearRegression(tau=0.1)
model.fit(X, y)

# Predict y_query
y_query = model.predict(x_query)

# Plot the data
model.plot(X, y, x_query, y_query)
```
