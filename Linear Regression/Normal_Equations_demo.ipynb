#### Implement Linear Regression Model using Normal Equations on Synthetic Data

```python
import numpy as np
import pandas as pd
from sklearn.metrics import r2_score
from NE import NormalEquations
```

```python
# Generate synthetic training set
np.random.seed(42)
n_samples = 100
noise = np.random.randn(n_samples) * 15

# Generate y = 3X + 4 + noise
X = np.linspace(1, n_samples, n_samples)
y = 3 * X + 4 + noise
```

```python
# Fit the model
model = NormalEquations()
model.fit(X, y)
pred_y = model.predict(X)

# Plot the data
model.plot(X, y, pred_y)
```

#### Implement Linear Regression Model using Normal Equations on Predicting House Price

```python
# Load Data
house_data = pd.read_csv('../dataset/house_price.csv')

# Select first 100 training and next 50 testing data
train_X = np.array(house_data['sqft_living'].head(100))
train_Y = np.array(house_data['price'].head(100))
test_X = np.array(house_data['sqft_living'].iloc[100:200])
test_Y = np.array(house_data['price'].iloc[100:200])
```

```python
# Fit the model
model = NormalEquations()
model.fit(train_X, train_Y)
pred_Y = model.predict(train_X)

# Plot the data
model.plot(train_X, train_Y, pred_Y)
```

```python
# Evaluate model performance on test data
pred_Y = model.predict(test_X)

# Calculate R-squared
r2 = r2_score(test_Y, pred_Y)
print("R-squared score (R): ", r2)
```
