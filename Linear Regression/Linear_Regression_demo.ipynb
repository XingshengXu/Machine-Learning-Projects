#### Implement Linear Regression using Batch Gradient Descent (BGD) on Synthetic Data

```python
import numpy as np
import pandas as pd
from Linear_Regression import LinearRegression
from Linear_Regression_utils import *
```

```python
%matplotlib widget

# Generate synthetic training set
np.random.seed(42)
n_samples = 100
noise = np.random.randn(n_samples) * 15

# Generate y = 3X + 4 + noise
X = np.linspace(1, n_samples, n_samples)
y = 3 * X + 4 + noise
```

```python
# Fit the model
model = LinearRegression('BGD', learning_rate=0.1, max_iterations=100)
model.fit(X, y)
```

```python
# Plot cost vs. iteration
plot_cost_vs_iteration(model)
```

#### Implement Linear Regression using Batch Gradient Descent (BGD) on Predicting House Price with Performance Test

```python
# Load Data
house_data = pd.read_csv('../dataset/house_price.csv')

# Select first 100 training and next 50 testing data
train_X = np.array(house_data['sqft_living'].head(100))
train_Y = np.array(house_data['price'].head(100))
test_X = np.array(house_data['sqft_living'].iloc[100:200])
test_Y = np.array(house_data['price'].iloc[100:200])
```

```python
# Fit the model
model = LinearRegression('BGD', learning_rate=0.1, max_iterations=100)
model.fit(train_X, train_Y)
```

```python
# Plot cost vs. iteration
plot_cost_vs_iteration(model)

# Evaluate model performance on training data
pred_y = [model.predict_value(x) for x in test_X]
evaluate_model(model, test_Y, pred_y)
```

#### Implement Linear Regression using Stochastic Gradient Descent (SGD) on Predicting House Price with Performance Test

```python
# Fit the model
model = LinearRegression('SGD', learning_rate=0.01, max_iterations=100)
model.fit(train_X, train_Y)
```

```python
# Plot cost vs. iteration
plot_cost_vs_iteration(model)

# Evaluate model performance on test data
pred_y = [model.predict_value(x) for x in test_X]
evaluate_model(model, test_Y, pred_y)
```
