#### Implement Logistic Regression on Synthetic Data

##### 1. Logistic regression using batch gradient descent on linearly separable binary-class data

```python
X,y = make_blobs(n_samples=100, n_features=2, centers=2, cluster_std=0.2, random_state=0)

# Fit the model
model = LogisticRegression(algorithm='BGD')
model.fit(X, y)

# Plot the data
model.create_contour_plot(X, y)
plot_cost_vs_iteration(model)
```

#### Implement Logistic Regression on Marketing Target Data

##### 1. Logistic regression using batch gradient descent on marketing target data with performance test

```python
# Load Data
market_data = pd.read_csv('../dataset/Social_Network_Ads.csv', header=0)

# Remove the First Column
market_data = market_data.drop(market_data.columns[0], axis=1)

# Select first 300 training and 100 testing data
# Training Set
train_X = np.array(market_data.iloc[0:300, 0:-1])
train_Y = np.array(market_data.iloc[0:300:, -1])

# Test Set
test_X = np.array(market_data.iloc[300:, 0:-1])
test_Y = np.array(market_data.iloc[300:, -1])
```

```python
# Fit the model
model = LogisticRegression(algorithm='BGD')
model.fit(train_X, train_Y)

pred_y = model.predict_class(test_X)

# Plot the data
model.create_contour_plot(train_X, train_Y)
plot_cost_vs_iteration(model)

# Performance Test
evaluate_model(test_Y, pred_y)
```

##### 2. Logistic regression using newton method on marketing target data with performance test

```python
# Fit the model
model = LogisticRegression(algorithm='NM')
model.fit(train_X, train_Y)

pred_y = model.predict_class(test_X)

# Plot the data
model.create_contour_plot(train_X, train_Y)
plot_cost_vs_iteration(model)

# Performance Test
evaluate_model(test_Y, pred_y)
```
