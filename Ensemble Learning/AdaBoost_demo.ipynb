#### Implement AdaBoost Classifier on Synthetic Data

```python
import numpy as np
import pandas as pd
import path_setup
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
from Ensemble_Learning import AdaBoostClassifier
from ensemble_utils import *
```

##### 1. Implement AdaBoost Classifier on manually generated two-class data

##### 2. Implement AdaBoost Classifier on non-linearly separable two-class data

```python
# Create training set
X,y = make_blobs(n_samples=100, n_features=2, centers=2, cluster_std=2, random_state=0)

# Split the data into training and test sets
train_X, test_X, train_Y, test_Y = train_test_split(X, y, test_size=0.3, random_state=0)

# Create ensemble model
abc = AdaBoostClassifier(model_number=10, learning_rate=0.1)
abc.fit(train_X, train_Y)

# Performance Test
pred_y = abc.predict_class(test_X)
evaluate_model(test_Y, pred_y, 'classification')

# Plot decision boundaries and class labels
print('Decision Boundary with Training Data')
create_contour_plot(abc, train_X, train_Y, resolution=200)

print('Decision Boundary with Testing Data')
create_contour_plot(abc, test_X, test_Y, resolution=200)
```

#### Implement AdaBoost Classifier on Predicting Marketing Target

```python
# Load Data
market_data = pd.read_csv('../dataset/Social_Network_Ads.csv', header=0)

# Remove the First Column
market_data = market_data.drop(market_data.columns[0], axis=1)

# Training Set
train_X = market_data.iloc[0:300, 0:-1].values
train_Y = market_data.iloc[0:300:, -1].values

# Test Set
test_X = market_data.iloc[300:, 0:-1].values
test_Y = market_data.iloc[300:, -1].values
```

```python
# Create ensemble model
abc = AdaBoostClassifier(model_number=10, learning_rate=0.1)
abc.fit(train_X, train_Y)

# Performance Test
pred_y = abc.predict_class(test_X)
evaluate_model(test_Y, pred_y, 'classification')

# Plot decision boundaries and class labels
print('Decision Boundary with Training Data')
create_contour_plot(abc, train_X, train_Y, resolution=200)

print('Decision Boundary with Testing Data')
create_contour_plot(abc, test_X, test_Y, resolution=200)
```
