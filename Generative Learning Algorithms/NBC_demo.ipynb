#### Naive Bayes Classifier for Classificating Spam Emails

```python
import pandas as pd
from Naive_Bayes_Classifier import NaiveBayesClassifier
from GLA_utils import *
```

```python
# Settings
train_data_size = 5000

# Load Data
email_data = pd.read_csv('../dataset/emails.csv', header=0)

# Remove the First Column
email_data.drop(email_data.columns[0], axis=1, inplace=True)
for col in email_data.columns:
    email_data.loc[email_data[col] > 0, col] = 1

# Training Set
train_X = email_data.iloc[0:train_data_size, 0:-1].values
train_Y = email_data.iloc[0:train_data_size:, -1].values

# Test Set
test_X = email_data.iloc[train_data_size:, 0:-1].values
test_Y = email_data.iloc[train_data_size:, -1].values
```

```python
# Fit the model
nbc = NaiveBayesClassifier(algorithm='NBC')
nbc.fit(train_X, train_Y)

pred_y = nbc.predict_class(test_X)

# Performance Test
evaluate_model(test_Y, pred_y)
```

#### Multinomial Naive Bayes Classifier for Classificating Spam Emails

```python
# Load Data
email_data = pd.read_csv('../dataset/emails.csv', header=0)

# Remove the First 1000 Columns of High Frequency Words
email_data.drop(email_data.columns[:1000], axis=1, inplace=True)

# Settings
train_data_size = 5000

# Training Set
train_X = email_data.iloc[0:train_data_size, 0:-1].values
train_Y = email_data.iloc[0:train_data_size:, -1].values

# Test Set
test_X = email_data.iloc[train_data_size:, 0:-1].values
test_Y = email_data.iloc[train_data_size:, -1].values
```

```python
# Fit the model
nbc = NaiveBayesClassifier(algorithm='MNBC')
nbc.fit(train_X, train_Y)

pred_y = nbc.predict_class(test_X)

# Performance Test
evaluate_model(test_Y, pred_y)
```
