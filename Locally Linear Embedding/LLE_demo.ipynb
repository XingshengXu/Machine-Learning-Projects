#### Locally Linear Embedding Demos

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_blobs
from Locally_Linear_Embedding import *
```

##### 1. Implement Locally Linear Embedding on non-linearly separable two-class data

```python
# Create dataset
X, y = make_blobs(n_samples=200, n_features=3, centers=2,
                  cluster_std=2, random_state=0)

# Apply LLE to the dataset
lle = LocallyLinearEmbedding()
transform_X = lle.fit_transform(X)

# Plot the original input data in 3D
plot_3d_data(X, y)

# Plot transformed data
plot_lle_embedding(transform_X, y)

# Plot the eigenvalues
eigenvalues = lle.eigenvalues 
plt.bar(range(len(eigenvalues)), eigenvalues)
plt.xlabel('Eigenvalue Index')
plt.ylabel('Eigenvalue')
plt.title('LLE Eigenvalues Plot')
plt.show()
```

##### 2. Implement Locally Linear Embedding on swiss roll data

```python
# Create dataset
t = np.arange(0, 10, 0.01)
x0 = np.cos(t) * (t + 1)
x2 = np.sin(t) * (t + 1)
x1 = 20 * np.random.rand(t.shape[0])

X = np.column_stack([x0, x1, x2])

# Apply LLE to the dataset
lle = LocallyLinearEmbedding(k_neighbors=10)
transform_X = lle.fit_transform(X)

# Plot the original input data in 3D
plot_3d_data(X, y=None, t=t, rotate=False)
plot_3d_data(X, y=None, t=t, rotate=True)

# Plot transformed data
plot_lle_embedding(transform_X, y=None, t=t)
```

##### 3. Implement Locally Linear Embedding on arrow shape data

```python
# Create dataset
t = np.arange(-1, 1, 0.001)
x0 = t

# Vectorized computation for x1
mask1 = (t < 0.5) & (t > -0.5)
mask2 = t <= -0.5
mask3 = ~mask1 & ~mask2

x1 = np.zeros_like(t)
x1[mask1] = 0.02 * np.random.uniform(-1, 1, mask1.sum())
x1[mask2] = 2 * (1 + t[mask2]) * np.random.uniform(-1, 1, mask2.sum())
x1[mask3] = 2 * (-0.5 + t[mask3]) * np.random.uniform(-1, 1, mask3.sum())

x2 = np.cos(t*5 + 1.5)

X = np.column_stack([x0, x1, x2])

# Apply LLE to the dataset
lle = LocallyLinearEmbedding(k_neighbors=30)
transform_X = lle.fit_transform(X)

# Plot the original input data in 3D
plot_3d_data(X, y=None, t=t, rotate=False)
plot_3d_data(X, y=None, t=t, rotate=True)

# Plot transformed data
plot_lle_embedding(transform_X, y=None, t=t)
```
