#### Implement Neural Perceptron Classifier for Handwritten Digit Recognition

##### 1. Single layer perceptron on handwritten digit recognition

```python
import idx2numpy
import pandas as pd
from sklearn.preprocessing import StandardScaler
from Single_Layer_Perceptron import SLPClassifier
from Multi_Layer_Perceptron import MLPClassifier
from Perceptron_utils import *
```

```python
# Load Training and Testing Data Sets
train_X = idx2numpy.convert_from_file(
    '../dataset/train-images.idx3-ubyte')
train_Y = idx2numpy.convert_from_file(
    '../dataset/train-labels.idx1-ubyte')
test_X = idx2numpy.convert_from_file(
    '../dataset/t10k-images.idx3-ubyte')
test_Y = idx2numpy.convert_from_file(
    '../dataset/t10k-labels.idx1-ubyte')
```

```python
# Preprocess the image into input data format
train_X = preprocess_image(train_X)
test_X = preprocess_image(test_X)

# Fit the model
slp = SLPClassifier(learning_rate=0.00001, max_iterations=1000)
slp.fit(train_X, train_Y)

pred_y = slp.predict_class(test_X)

# Plot the data
plot_cost_vs_iteration(slp)

# Performance Test
evaluate_model(test_Y, pred_y)
```

##### 2. Single layer perceptron on predicting Marketing Target

```python
# Load Data
market_data = pd.read_csv('../dataset/Social_Network_Ads.csv', header=0)

# Remove the First Column
market_data = market_data.drop(market_data.columns[0], axis=1)

# Training Set
train_X = market_data.iloc[0:300, 0:-1].values
train_Y = market_data.iloc[0:300:, -1].values

# Test Set
test_X = market_data.iloc[300:, 0:-1].values
test_Y = market_data.iloc[300:, -1].values

# Standardize the Features for Both Training Set and Test Set
scaler = StandardScaler()

train_X = scaler.fit_transform(train_X)
test_X = scaler.transform(test_X)
```

```python
# Fit the model
slp = SLPClassifier(learning_rate=0.02, max_iterations=100)
slp.fit(train_X, train_Y)

# Plot the data
plot_cost_vs_iteration(slp)

# Performance Test
pred_y = slp.predict_class(test_X)
evaluate_model(test_Y, pred_y)

# Plot the decision boundary and the true labels
print('Decision Boundary with Training Data')
create_contour_plot(slp, train_X, train_Y, resolution=200)

# Plot the decision boundary and the true labels
print('Decision Boundary with Testing Data')
create_contour_plot(slp, test_X, test_Y, resolution=200)
```

##### 3. Multi layer perceptron on handwritten digit recognition

```python
# Load Training and Testing Data Sets
train_X = idx2numpy.convert_from_file(
    '../dataset/train-images.idx3-ubyte')
train_Y = idx2numpy.convert_from_file(
    '../dataset/train-labels.idx1-ubyte')
test_X = idx2numpy.convert_from_file(
    '../dataset/t10k-images.idx3-ubyte')
test_Y = idx2numpy.convert_from_file(
    '../dataset/t10k-labels.idx1-ubyte')
```

```python
# Preprocess the image into input data format
train_X = preprocess_image(train_X)
test_X = preprocess_image(test_X)

# Fit the model
mlp = MLPClassifier(learning_rate=0.1, hidden_node_number=10, max_iterations=1000)
mlp.fit(train_X, train_Y)

pred_y = mlp.predict_class(test_X)

# Plot the data
plot_cost_vs_iteration(mlp)

# Performance Test
evaluate_model(test_Y, pred_y)
```

##### 4. Multi layer perceptron on predicting Marketing Target

```python
# Load Data
market_data = pd.read_csv('../dataset/Social_Network_Ads.csv', header=0)

# Remove the First Column
market_data = market_data.drop(market_data.columns[0], axis=1)

# Training Set
train_X = market_data.iloc[0:300, 0:-1].values
train_Y = market_data.iloc[0:300:, -1].values

# Test Set
test_X = market_data.iloc[300:, 0:-1].values
test_Y = market_data.iloc[300:, -1].values

# Standardize the Features for Both Training Set and Test Set
scaler = StandardScaler()

train_X = scaler.fit_transform(train_X)
test_X = scaler.transform(test_X)
```
