#### Maximum Variance Unfolding Demos

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_blobs
from Maximum_Variance_Unfolding import *
```

##### 1. Implement Maximum Variance Unfolding on non-linearly separable two-class data

```python
# Create dataset
X, y = make_blobs(n_samples=200, n_features=3, centers=2,
                  cluster_std=2, random_state=0)

# Apply MVU to the dataset
mvu = MaximumVarianceUnfolding(k_neighbors=30)
transform_X = mvu.fit_transform(X)

# Plot the original input data in 3D
plot_3d_data(X, y)

# Plot transformed data
plot_mvu_embedding(transform_X, y)
```

##### 2. Implement Maximum Variance Unfolding on swiss roll data

```python
# Create dataset
t = np.arange(0, 10, 0.01)
x0 = np.cos(t) * (t + 1)
x2 = np.sin(t) * (t + 1)
x1 = 10 * np.random.rand(t.shape[0])

X = np.column_stack([x0, x1, x2])

# Apply MVU to the dataset
mvu = MaximumVarianceUnfolding(k_neighbors=60)
transform_X = mvu.fit_transform(X)

# Plot the original input data in 3D
plot_3d_data(X, y=None, t=t, rotate=False)
plot_3d_data(X, y=None, t=t, rotate=True)

# Plot transformed data
plot_mvu_embedding(transform_X, y=None, t=t)
```

##### 3. Implement Maximum Variance Unfolding on arrow shape data

```python
# Create dataset
t = np.arange(-1, 1, 0.001)
x0 = t

# Vectorized computation for x1
mask1 = (t < 0.5) & (t > -0.5)
mask2 = t <= -0.5
mask3 = ~mask1 & ~mask2

x1 = np.zeros_like(t)
x1[mask1] = 0.02 * np.random.uniform(-1, 1, mask1.sum())
x1[mask2] = 2 * (1 + t[mask2]) * np.random.uniform(-1, 1, mask2.sum())
x1[mask3] = 2 * (-0.5 + t[mask3]) * np.random.uniform(-1, 1, mask3.sum())

x2 = np.cos(t*5 + 1.5)

X = np.column_stack([x0, x1, x2])

# Apply MVU to the dataset
mvu = MaximumVarianceUnfolding(k_neighbors=150)
transform_X = mvu.fit_transform(X)

# Plot the original input data in 3D
plot_3d_data(X, y=None, t=t, rotate=False)
plot_3d_data(X, y=None, t=t, rotate=True)

# Plot transformed data
plot_mvu_embedding(transform_X, y=None, t=t)
```

##### 4. Implement Maximum Variance Unfolding on helix shape data

```python
# Create dataset
t = np.linspace(0, 4 * np.pi, 1000)
x = np.sin(t)
y = np.cos(t)
z = t

# Add noise
x += 0.05 * np.random.randn(x.shape[0])
y += 0.05 * np.random.randn(y.shape[0])

X = np.column_stack([x, y, z])

# Apply MVU to the dataset
mvu = MaximumVarianceUnfolding(k_neighbors=50)
transform_X = mvu.fit_transform(X)

# Plot the original input data in 3D
plot_3d_data(X, y=None, t=t, rotate=False)
plot_3d_data(X, y=None, t=t, rotate=True)

# Plot transformed data
plot_mvu_embedding(transform_X, y=None, t=t)
```

##### 5. Implement Maximum Variance Unfolding on twin peaks data

```python
# Create dataset
def generate_peak(t, center, std_dev):
    return np.exp(-(t - center)**2 / (2 * std_dev**2))

t = np.linspace(-1, 1, 1000)
x = generate_peak(t, 0.3, 0.1) + generate_peak(t, -0.3, 0.1)
x += 0.003 * np.random.randn(t.size)
y = t
z = generate_peak(t, 0.5, 0.15) + generate_peak(t, -0.5, 0.15)
z += 0.003 * np.random.randn(t.size)

X = np.column_stack([x, y, z])

# Apply MVU to the dataset
mvu = MaximumVarianceUnfolding(k_neighbors=80)
transform_X = mvu.fit_transform(X)

# Plot the original input data in 3D
plot_3d_data(X, y=None, t=t, rotate=False)
plot_3d_data(X, y=None, t=t, rotate=True)

# Plot transformed data
plot_mvu_embedding(transform_X, y=None, t=t)
```
